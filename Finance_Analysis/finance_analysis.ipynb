```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("stock_market_data.csv")

# Display first 5 rows
df.head()
```

```python
# Dataset shape
df.shape
```

## Dataset Overview :

This dataset contains historical stock market data including company symbols, trading dates, opening prices, closing prices, highest and lowest prices of the day, and trading volume.

The dataset enables analysis of stock price trends, market volatility, and trading activity over time.

Such financial data is commonly used by analysts to understand market behavior, identify patterns, and generate investment insights.

```python
# Dataset information
df.info()
```

##Data Quality Assessment :

The dataset contains 1862 records and 10 features representing financial market indicators over time.

All columns contain complete data with no missing values, indicating strong data quality and eliminating the need for imputation or missing value handling.

Most variables are numerical financial indicators including stock market index values, dividends, earnings, inflation indicators, and interest rates. The dataset also includes a Date column representing the time dimension for the analysis.

This structure allows for time-series based financial analysis and exploration of relationships between market valuation metrics and macroeconomic indicators.

```python
# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Verify datatype
df["Date"].dtype
```

##Date Conversion :

The Date column was converted from object format to datetime format to enable time-based analysis. This allows the dataset to be used for trend analysis, time-series visualization, and long-term financial pattern exploration.

```python
df.describe()
```

##Statistical Summary :

The statistical summary provides an overview of the financial indicators present in the dataset.

The S&P 500 index values show significant variation across the time period, reflecting long-term market growth and fluctuations.

Macroeconomic indicators such as the Consumer Price Index and Long Interest Rate provide additional economic context that may influence stock market valuation.

Valuation metrics such as the PE10 ratio help evaluate whether the market is relatively overvalued or undervalued during different periods.

```python
plt.figure()
plt.plot(df["Date"], df["SP500"])
plt.xlabel("Year")
plt.ylabel("S&P 500 Index Value")
plt.title("S&P 500 Market Trend Over Time")
plt.show()
```

##Market Trend Analysis :

The visualization shows the long-term movement of the S&P 500 index over time.

The general upward trend indicates long-term growth in the stock market despite periods of volatility and economic downturns.

Such patterns highlight the long-term value creation potential of equity markets while also demonstrating the cyclical nature of financial markets.

```python
plt.figure()
plt.plot(df["Date"], df["Dividend"])
plt.xlabel("Year")
plt.ylabel("Dividend")
plt.title("Dividend Trend Over Time")
plt.show()
```

##Dividend Trend Analysis :

The dividend trend shows how corporate dividend payouts have changed over time. Dividends represent a portion of company profits returned to shareholders.

A stable or increasing dividend trend often indicates financial stability and strong corporate earnings. Investors commonly analyze dividend trends to evaluate income-generating investment opportunities and long-term company performance.

```python
plt.figure()
plt.plot(df["Date"], df["Earnings"])
plt.xlabel("Year")
plt.ylabel("Earnings")
plt.title("Corporate Earnings Trend Over Time")
plt.show()
```

##Earnings Trend Analysis :

Corporate earnings represent the profitability of companies included in the market index.

The earnings trend provides insight into the financial health of corporations over time. Growth in earnings typically supports higher stock market valuations, while declining earnings may signal economic slowdowns or reduced business performance.

Understanding earnings trends helps investors evaluate long-term market sustainability.

```python
correlation = df.drop(columns=["Date"]).corr()

plt.figure()
sns.heatmap(correlation, annot=True)
plt.title("Financial Indicator Correlation Heatmap")
plt.show()
```

##Correlation Analysis of Financial Indicators :

The correlation heatmap illustrates the relationships between key financial indicators in the dataset.

Strong positive correlations between variables such as stock prices, earnings, and dividends suggest that market valuation is closely tied to corporate profitability.

Macroeconomic indicators like interest rates and inflation may also influence market behavior. Understanding these relationships helps analysts identify key drivers behind financial market movements.

Macroeconomic indicators such as interest rates and inflation may also influence market conditions, but corporate earnings remain a key determinant of long-term market behavior.
